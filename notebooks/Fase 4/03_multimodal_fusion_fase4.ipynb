
## Fase 4 — Fusão Multimodal para Identificação de Vulnerabilidade Emocional



### Objetivo

Este notebook tem como objetivo consolidar os resultados obtidos nas
etapas anteriores de análise de vídeo e análise de áudio.

A proposta consiste em realizar fusão multimodal entre indicadores
visuais e emocionais, permitindo geração de um score consolidado
de risco e classificação final da análise.



### Fundamentação Teórica

A multimodalidade em Inteligência Artificial consiste na combinação
de múltiplas fontes de informação para melhorar interpretação
e contextualização dos resultados.

Neste projeto, a fusão multimodal combina:
- análise visual baseada em YOLOv8;
- análise de áudio baseada em Azure Speech Services;
- análise de sentimento utilizando Azure AI Language.

A combinação dessas modalidades permite geração de indicadores
mais completos dentro do pipeline proposto pelo desafio.



### Motivação da Abordagem

O documento do Tech Challenge solicita explicitamente a utilização
de IA multimodal aplicada ao contexto de saúde da mulher.

Dessa forma, foi implementada uma abordagem simplificada de fusão
multimodal baseada em combinação heurística de scores provenientes
das modalidades de vídeo e áudio.



### Importação das Bibliotecas


```python

import json
import matplotlib.pyplot as plt
import pandas as pd
```


### Simulação dos Resultados das Etapas Anteriores

Nesta etapa são carregados os resultados gerados pelos notebooks
de vídeo e áudio.

Os dados representam:
- score visual de risco;
- score emocional de áudio;
- indicadores emocionais;
- análise comportamental.


```python

video_report = {
    "video_risk_score": 0.72,
    "movement_pattern": "agitated",
    "posture_analysis": "defensive posture"
}

audio_report = {
    "audio_risk_score": 0.81,
    "sentiment": "negative",
    "key_phrases": [
        "muito cansada",
        "aperto no peito",
        "preocupada"
    ]
}

print("Video Report")
print(json.dumps(video_report, indent=4))

print("\nAudio Report")
print(json.dumps(audio_report, indent=4))
```


### Fusão Multimodal

Nesta etapa é realizado o processo de fusão entre os scores
gerados pelas modalidades de vídeo e áudio.

Foi utilizada uma abordagem heurística simples baseada
em média ponderada entre os indicadores multimodais.


```python

VIDEO_WEIGHT = 0.5
AUDIO_WEIGHT = 0.5

final_risk_score = (
    video_report["video_risk_score"] * VIDEO_WEIGHT
    + audio_report["audio_risk_score"] * AUDIO_WEIGHT
)

print(f"Final Risk Score: {final_risk_score:.2f}")
```


### Classificação Final de Risco

Após a fusão multimodal, o score consolidado é utilizado
para classificação final do nível de risco.


```python

if final_risk_score < 0.3:

    risk_level = "LOW"

elif final_risk_score < 0.7:

    risk_level = "MEDIUM"

else:

    risk_level = "HIGH"

print(f"Risk Level: {risk_level}")
```


### Geração de Interpretação Consolidada

A interpretação consolidada é gerada utilizando os resultados
das modalidades de vídeo e áudio.


```python

interpretation = (
    f"Multimodal analysis identified {risk_level.lower()} "
    f"risk indicators based on emotional and behavioral patterns. "
    f"Relevant audio indicators include: "
    f"{', '.join(audio_report['key_phrases'])}."
)

print(interpretation)
```


### Visualização dos Scores Multimodais


```python

scores_df = pd.DataFrame({
    "modality": [
        "Video",
        "Audio",
        "Final"
    ],
    "score": [
        video_report["video_risk_score"],
        audio_report["audio_risk_score"],
        final_risk_score
    ]
})

plt.figure(figsize=(8, 5))

plt.bar(
    scores_df["modality"],
    scores_df["score"]
)

plt.ylim(0, 1)

plt.title("Scores Multimodais")
plt.xlabel("Modalidade")
plt.ylabel("Score")

plt.show()
```


### Geração de Relatório Consolidado

Nesta etapa é gerado o relatório final consolidando:
- resultados visuais;
- resultados emocionais;
- score multimodal;
- classificação final;
- interpretação contextual.


```python

multimodal_report = {
    "video_risk_score": video_report["video_risk_score"],
    "audio_risk_score": audio_report["audio_risk_score"],
    "final_risk_score": final_risk_score,
    "risk_level": risk_level,
    "key_phrases": audio_report["key_phrases"],
    "interpretation": interpretation
}

print(json.dumps(multimodal_report, indent=4))
```


### Interpretação dos Resultados

Os resultados demonstram a viabilidade da utilização de técnicas
de fusão multimodal para consolidação de informações provenientes
de diferentes modalidades de IA.

A combinação entre análise visual e análise emocional permitiu
geração de indicadores integrados de risco dentro do contexto
proposto pelo desafio.
