# Contexto do Problema

## O objetivo deste projeto é apoiar o diagnóstico de diabetes mellitus a partir de dados clínicos estruturados, utilizando modelos de aprendizado de máquina supervisionado.
## Na Fase 1, diferentes algoritmos de classificação foram avaliados, e a Regressão Logística apresentou o melhor equilíbrio entre:

### - desempenho (especialmente recall da classe positiva),
### - simplicidade
### - interpretabilidade — fator essencial no contexto médico.

## Nesta fase, o foco passa a ser a otimização dos hiperparâmetros desse modelo por meio de Algoritmos Genéticos (GA).
